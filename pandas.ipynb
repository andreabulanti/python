```python
#panoramica dataset
dfdiab.info(memory_usage="deep")
```

```python
#aggiunta colonna condizionale per raggruppare le date
condizioni = [
    (dfdiab["Age"] < 20),
    (dfdiab["Age"] >= 20) & (dfdiab["Age"] < 30),
    (dfdiab["Age"] >= 30) & (dfdiab["Age"] < 40),
    (dfdiab["Age"] >= 40) & (dfdiab["Age"] < 50),
    (dfdiab["Age"] >=50)
]

valori = ["<20", "20-30", "30-40", "40-50",">50"]

# Applica le condizioni utilizzando numpy.select() per creare la nuova colonna
dfdiab["Range_age"] = np.select(condizioni, valori, default="Altra_condizione")
dfdiab
```

```python
#media della pressione sanguigna
dfdiab.groupby("Range_age", as_index=False).agg({"BloodPressure": "mean"})
```

## Insurance

```python
dfins = pd.read_csv("insurance.csv")
dfins.head()
```

```python
#no empty value
dfins.isna().sum()
```

```python
#media charges per region
(dfins.groupby("region")[["charges"]].mean()
     .sort_values(by="charges", ascending=False)
     .plot.bar(cmap="Pastel2")
);
```

```python
(dfins.groupby("smoker")[["charges"]].mean()
     .sort_values(by="charges", ascending=False)
     .plot.bar(cmap="flag")
);
```

```python
(dfins.groupby("sex")[["charges"]].mean()
     .sort_values(by="charges", ascending=False)
     .plot.bar(cmap="plasma")
);
```

```python
#la media dei valori di bmi si colloca fra 25 e 35 bmi
dfins["bmi"].plot.hist(cmap="Set3", alpha = 0.4)
```

```python
bmi_min = dfins["bmi"].min() 
bmi_max = dfins["bmi"].max()
bmi_min, bmi_max
```

```python
#troviamo i quartili
q25, q50, q75 = np.percentile(dfins["bmi"], [25, 50, 75])
iqr = q75 - q25
```

```python
min_bin = q25 - 1.5*iqr
max_bin = q50 + 1.5*iqr
```

```python
print("Il valore minimo dei bin è pari a ",bmi_min)
print("Il valore massimo dei bin è pari a ",bmi_max)
print(f"I quartili sono distribuiti da Q1 a Q3 nel seguente modo: {q25}, {q50}, {q75}")
```

```python
print(f"il valore massimo oltre il quale è probabile trovare degli oulier è pari a {max_bin}")
print(f"il valore minimo oltre il quale è probabile trovare degli oulier è pari a {min_bin}")
```

```python
dfins["bmi"].plot.box()
```

```python
outmask = (dfins["bmi"]>max_bin) | (dfins["bmi"] < min_bin)
dfins.loc[outmask]
```

```python
#pulizia del dataset: rimozione degli outliers
clean_df = dfins.loc[~outmask]
clean_df["bmi"].plot.hist(cmap = "Accent", alpha = 0.3)
```
